```python
import pandas as pd
import sys

sys.path.append('..')
from modules.plots import gboxplot
```

```python
path = "./results"
folders = [
    ('SM_1H_K2',11,"CSNN"),
    ('MM_1H_K2',15,"CSNN"),
    ('LM_1H_K2',23,"CSNN"),
    ('SM_2H_K3',16,"CSNN"),
    ('MM_2H_K2',9,"CSNN"),
    ('LM_2H_K3',23,"CSNN"),

    ('SM_3H_K2',21,"CSNN"),
    ('MM_3H_K2',22,"CSNN"),
    ('LM_3H_K2',10,"CSNN"),

    ("FFSNN_1H",22,"FFSNN"),
    ("FFSNN_2H",25,"FFSNN"),
    ("FFSNN_3H",21,"FFSNN"),

    ("CNN_1H",1,"CNN"),
    ("CNN_2H",1,"CNN"),
    ("CNN_3H",1,"CNN"),

    ("FFNN_1H",1,"FFNN"),
    ("FFNN_2H",1,"FFNN"),
    ("FFNN_3H",1,"FFNN"),
]

dfs = []
for f, steps,type in folders:
    results_df = pd.read_csv(f'{path}/{f}.csv')
    model = f.split('-')[0]
    results_df['model'] = model
    results_df['steps'] = steps
    results_df['type'] = type
    dfs.append(results_df)
    print(f'Loaded {f} – {len(results_df)} results')
len(dfs)
```

```python
df = pd.concat(dfs)
print(df.shape)
df
```

```python
df_metrics = df[df['recall@5FPR'] > 0].copy()
metrics = ['Accuracy', 'F1-Score', 'Precision', 'Recall']
df_metrics[metrics[0]] = df_metrics['accuracy']
df_metrics[metrics[1]] = df_metrics['f1_score']
df_metrics[metrics[2]] = df_metrics['precision']
df_metrics[metrics[3]] = df_metrics['recall']
df_metrics = df_metrics.melt(id_vars=['model', 'dataset'], value_vars=metrics)
df_metrics = df_metrics[df_metrics['value'] > 0]
gboxplot(df_metrics, 
        figsize=(10, 3.5),
        x='variable', 
        y='value', 
        z='model',
        x_name='Metric',
        y_name='Value',
        z_name='Model',
        loc='upper center',
        ncol=6,
        bbox_to_anchor=(0.5, 1.35),
)
```

```python
df_performance = df[df['recall@5FPR'] > 0].copy()
metrics = ['TPR$_{5\%FPR}$', 'PE$_{age}$', 'PE$_{income}$', 'PE$_{employment}$']
df_performance[metrics[0]] = df_performance['recall@5FPR']
df_performance[metrics[1]] = df_performance['fpr_ratio_age']
df_performance[metrics[2]] = df_performance['fpr_ratio_income']
df_performance[metrics[3]] = df_performance['fpr_ratio_employment']
df_performance = df_performance.melt(id_vars=['model', 'dataset'], value_vars=metrics)
df_performance = df_performance[df_performance['value'] > 0]
gboxplot(df_performance, 
        figsize=(10, 3.5),        
        x='variable', 
        y='value', 
        z='model',
        x_name='Metric',
        y_name='Value',
        z_name='Model',
        loc='upper center',
        ncol=6,
        bbox_to_anchor=(0.5, 1.35),
)
```

```python
df_energy = df[df['recall@5FPR'] > 0].copy()
df_energy['joules'] = df_energy['energy_test']
df_energy['kilojoules'] = df_energy['energy_test'] / 1000
df_energy['watts_hours'] = df_energy['joules'] / 3600
df_energy['watts'] = df_energy['joules'] / df_energy['time_test']
df_energy['joules/steps'] = df_energy['joules'] / df_energy['steps']
df_energy['watts/steps'] = df_energy['watts'] / df_energy['steps']
metrics = ['Joules $[J]$', 'Watts $[W]$', 'Joules per Step', 'Watts per Step']
df_energy[metrics[0]] = df_energy['joules']
df_energy[metrics[1]] = df_energy['watts']
df_energy[metrics[2]] = df_energy['joules/steps']
df_energy[metrics[3]] = df_energy['watts/steps']
df_energy = df_energy.melt(id_vars=['model', 'dataset'], value_vars=metrics)
df_energy = df_energy[df_energy['value'] > 0]
gboxplot(df_energy, 
        figsize=(10, 3.5),
        x='variable', 
        y='value', 
        z='model',
        x_name='Metric',
        y_name='Consumption',
        z_name='Model',
        loc='upper center',
        ncol=6,
        bbox_to_anchor=(0.5, 1.35),
        y_scale='log'
)
```
