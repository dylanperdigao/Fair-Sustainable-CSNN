```python
import pandas as pd
import os
import sys
sys.path.append('..')
```

```python
path = "./ecml2025-results"
folders = [
    ('SM_1H_K2',11),
    ('MM_1H_K2',15),
    ('LM_1H_K2',23),
    ('SM_2H_K3',16),
    ('MM_2H_K2',9),
    ('LM_2H_K3',23),
    ('SM_3H_K2',21),
    ('MM_3H_K2',22),
    ('LM_3H_K2',10),

    ("FFSNN_1H",22),
    ("FFSNN_2H",25),
    ("FFSNN_3H",21),

    ("CNN_1H",1),
    ("CNN_2H",1),
    ("CNN_3H",1),

    ("FFNN_1H",1),
    ("FFNN_2H",1),
    ("FFNN_3H",1),
]

dfs = []
for f, steps in folders:
    results_df = pd.read_csv(f'{path}/{f}.csv')
    model = f.split('-')[0]
    results_df['model'] = model
    results_df['steps'] = steps
    dfs.append(results_df)
    print(f'Loaded {f} – {len(results_df)} results')
len(dfs)
```

```python
df = pd.concat(dfs)
print(df.shape)
df
```

```python
df_energy = df[df['recall@5FPR'] > 0].copy()
df_energy['joules'] = df_energy['energy_test']
df_energy['watts_hours'] = df_energy['joules'] / 3600
df_energy['watts'] = df_energy['joules'] / df_energy['time_test']
df_energy['ratio_J'] = df_energy['joules'] / df_energy['steps']
df_energy['ratio_W'] = df_energy['watts'] / df_energy['steps']
df_energy = df_energy.groupby(['model']).agg({
    'time_test': ['mean', 'std'],
    'joules': ['mean', 'std'],
    'watts_hours': ['mean', 'std'],
    'watts': ['mean', 'std'],
    'ratio_J': ['mean', 'std'],
    'ratio_W': ['mean', 'std'],
})

df_formatted = pd.DataFrame(index=df_energy.index)
for metric in ['joules', 'watts', 'ratio_J', 'ratio_W']:
    df_formatted[metric] = (
        (df_energy[(metric, 'mean')]*1).round(2).astype(str) +
        ' $\\pm$ ' +
        (df_energy[(metric, 'std')]*1).round(2).astype(str)
    )

df_formatted.round(3)
```

```python
df_performance = df[df['recall@5FPR'] > 0].copy()
df_performance = df_performance.groupby(['model']).agg({
    'accuracy': ['mean', 'std'],
    'f1_score': ['mean', 'std'],
    'precision': ['mean', 'std'],
    'recall': ['mean', 'std'],
    'recall@5FPR': ['mean', 'std'],
    'fpr_ratio_age': ['mean', 'std'],
    'fpr_ratio_income': ['mean', 'std'],
    'fpr_ratio_employment': ['mean', 'std'],
})

df_formatted = pd.DataFrame(index=df_performance.index)
metrics = ['accuracy', 'f1_score', 'precision', 'recall']
metrics_constraint = ['recall@5FPR', 'fpr_ratio_age', 'fpr_ratio_income', 'fpr_ratio_employment']
for metric in metrics+metrics_constraint:
    df_formatted[metric] = (
        (df_performance[(metric, 'mean')]*100).round(2).astype(str) +
        ' $\\pm$ ' +
        (df_performance[(metric, 'std')]*100).round(2).astype(str)
    )

df_formatted.round(2)
```
