```python
import optuna
import pandas as pd

pd.set_option('display.max_columns', 100)
```

```python
configs = [
    ('SM_1H_K2', "CSNN"),
    ('SM_1H_K3', "CSNN"),
    ('SM_1H_K4', "CSNN"),
    ('SM_1H_K5', "CSNN"),
    ('MM_1H_K2', "CSNN"),
    ('MM_1H_K3', "CSNN"),
    ('MM_1H_K4', "CSNN"),
    ('MM_1H_K5', "CSNN"),
    ('LM_1H_K2', "CSNN"),
    ('LM_1H_K3', "CSNN"),
    ('LM_1H_K4', "CSNN"),
    ('LM_1H_K5', "CSNN"),

    ('SM_2H_K2', "CSNN"),
    ('SM_2H_K3', "CSNN"),
    ('SM_2H_K4', "CSNN"),
    ('MM_2H_K2', "CSNN"),
    ('MM_2H_K3', "CSNN"),
    ('MM_2H_K4', "CSNN"),
    ('LM_2H_K2', "CSNN"),
    ('LM_2H_K3', "CSNN"),
    ('LM_2H_K4', "CSNN"),

    ('SM_3H_K2', "CSNN"),
    ('MM_3H_K2', "CSNN"),
    ('LM_3H_K2', "CSNN"),

    ("FFSNN_1H", "OTH"),
    ("FFSNN_2H", "OTH"),
    ("FFSNN_3H", "OTH"),

    ("CNN_1H", "OTH"),
    ("CNN_2H", "OTH"),
    ("CNN_3H", "OTH"),

    ("FFNN_1H", "OTH"),
    ("FFNN_2H", "OTH"),
    ("FFNN_3H", "OTH"),
]
```

```python
best_trials = pd.DataFrame()
for study_name, db in configs:
    study = optuna.load_study(study_name=study_name, storage=f'sqlite:///ecml2025-{db}.db')
    df = study.trials_dataframe()
    df = df[df['values_0'] <= 0.05]
    print(df.shape, study_name)
    try:
        df = df.sort_values(by='user_attrs_@5FPR recall', ascending=False)
    except:
        print(f'{study_name} not found')
    df = df.head(1)
    df['net'] = study_name
    for i in range(1, 6):
        df[f'params_beta{i}'] = df[f'params_beta{i}'] if f'params_beta{i}' in df else None
        df[f'params_threshold{i}'] = df[f'params_threshold{i}'] if f'params_threshold{i}' in df else None
    df['params_slope'] = df['params_slope'] if 'params_slope' in df else None
    df['params_step'] = df['params_step'] if 'params_step' in df else None
    df['kernel'] = study_name.split('_')[-1] if db == "CSNN" else None
    df = df.dropna(axis=1, how='all')
    best_trials = pd.concat([best_trials, df])
best_trials = best_trials.sort_values(by='user_attrs_@5FPR recall', ascending=False)
best_trials
```

```python
best_hyperparameters = best_trials[[
    'net', 'params_epoch', 'params_batch', 'params_step',
    'params_slope',	
    'params_weight',
    'params_adam_beta1','params_adam_beta2',
    'params_learning_rate',	
    'user_attrs_@5FPR fpr', 'user_attrs_@5FPR recall', 
]]
best_parameters = best_trials[[
    'net',
    'params_beta1','params_beta2','params_beta3','params_beta4', 'params_beta5',
    'params_threshold1','params_threshold2','params_threshold3', 'params_threshold4', 'params_threshold5',
    'user_attrs_@5FPR fpr', 'user_attrs_@5FPR recall', 
]]
best_hyperparameters
```
